```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DISDATAV.csv')
df.head(10)
```

```python
df['HUM'] = df['HUM'].apply(lambda x: float(str(x)[0:5]))
df.head(10)
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.utils import shuffle
```

```python
df = shuffle(df)
df.head()
```

```python
X = df.drop(['TARGET'], axis = 1)
X = np.array(X)
Y = np.array(df['TARGET'])
X[:10], Y[:10]
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import to_categorical # Import directly from keras.utils
```

```python
l_encode = LabelEncoder()
l_encode.fit(Y)
Y = l_encode.transform(Y)
Y = to_categorical(Y)
Y
```

```python
Y.shape
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
scaled_X
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(scaled_X,Y, test_size = 0.3, random_state = 1)
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

```python
train_x = train_x.reshape((train_x.shape[0], train_x.shape[1], 1))
test_x = test_x.reshape((test_x.shape[0], test_x.shape[1], 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import ConvLSTM2D
import tensorflow as tf
```

```python
# Reshape train_x to include a timesteps dimension
train_x = train_x.reshape((train_x.shape[0], 1, train_x.shape[2], train_x.shape[3], train_x.shape[4]))

# Now define your model
model = Sequential()
model.add(ConvLSTM2D(100, kernel_size=(1,1), input_shape=(train_x.shape[1], train_x.shape[2], train_x.shape[3], train_x.shape[4]))) # input_shape is now 5D, reduced kernel size to (1,1)
model.add(Dropout(0.2))
model.add(Dense(train_y.shape[1], activation='softmax'))
```

```python
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Reshape train_x to include a timesteps dimension
train_x = train_x.reshape((train_x.shape[0], 1, train_x.shape[2], train_x.shape[3], train_x.shape[4]))

# Now define your model
model = Sequential()
model.add(ConvLSTM2D(100, kernel_size=(1,1), input_shape=(train_x.shape[1], train_x.shape[2], train_x.shape[3], train_x.shape[4]))) # input_shape is now 5D, reduced kernel size to (1,1)
model.add(Dropout(0.2))
# Add a Flatten layer to convert the 4D output of ConvLSTM2D to 2D
model.add(tf.keras.layers.Flatten())  # Flatten the output
model.add(Dense(train_y.shape[1], activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Then fit the model
callback = tf.keras.callbacks.EarlyStopping(monitor = 'loss', patience = 3)
history = model.fit(train_x, train_y, epochs = 50, batch_size = 32, callbacks = [callback], validation_data=(test_x, test_y))
```

```python
model.save("ConvLSTM2D.keras")  # Save the model with '.keras' extension
# or
model.save("ConvLSTM2D.h5")  # Save the model with '.h5' extension
```

```python
plt.rcParams['figure.dpi'] = 800
plt.plot(history.history['loss'])
plt.title('ConvLSTM2D model Train loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper right')
plt.savefig("ConvLSTM2D_model_train_loss.pdf", format="pdf", bbox_inches="tight")
plt.show()

# include Test loss in graph
# check for exclusive another for test loss
```

```python
plt.rcParams['figure.dpi'] = 800
plt.plot(history.history['val_loss'])
plt.title('ConvLSTM2D model Validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper right')
plt.savefig("ConvLSTM2D_model_val_loss.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
scores = model.evaluate(test_x, test_y)

for i, m in enumerate(model.metrics_names):
    print("\n%s: %.3f"% (m, scores[i]))
```

```python
predict_x=model.predict(test_x)
pred_=np.argmax(predict_x,axis=1)

true_y = l_encode.inverse_transform(np.argmax(to_categorical(test_y[:]), axis = 1)[:,1])

cnt = 0
for i,j in zip(pred_, true_y):
    print("Predicted: {}, True: {}".format(i, j))
    cnt = cnt+1
    if cnt > 100:
        break
```

```python
plt.rcParams['figure.dpi'] = 500
plt.plot(true_y, color = 'red', label = 'Real Class')
plt.plot(pred_, color = 'blue', label = 'Predicted Class')
plt.title('Classes of dataset')
plt.xlabel('Index')
plt.ylabel('Class')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
rmse = np.sqrt(mean_squared_error(true_y, pred_))
mse = mean_squared_error(true_y, pred_)
mae = mean_absolute_error(true_y, pred_)
r2 = r2_score(true_y, pred_)
print("TEST RMSE:", rmse)
print("TEST MSE:", mse)
print("TEST MAE:", mae)
print("TEST R2:", r2)
```
