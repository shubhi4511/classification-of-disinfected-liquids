```python
import numpy as np #to work with array, mathematical scientific calculation
import pandas as pd #to work on dataset
import matplotlib.pyplot as plt #visualization, poltting a graph, barchart, comparision chat, correlation matrix chart
```

```python
df = pd.read_csv('DISDATAV.csv') #load dataset
df.head(10) #loading 10 upper rows only
```

```python
df['HUM'] = df['HUM'].apply(lambda x: float(str(x)[0:5])) #convert string into float #df['HUM']: This selects the 'HUM' column from the DataFrame df.
df.head(10) #display 10 rows
```

```python
df.isnull().sum() #shows number of null values in each col
```

```python
for col in df.columns:
    df[col].fillna(df[col].mean(), inplace=True) #replace null value by its corresponding column
```

```python
df.isnull().sum()
```

```python
from sklearn.utils import shuffle #sklearn.utils is collection of utility function for manipulation
```

```python
df = shuffle(df)
df.head()#check after shuffling only 5 rows
```

```python
X = df.drop(['TARGET'], axis = 1) #drop target col to test it
X = np.array(X) #convert X into array
Y = np.array(df['TARGET']) # target col into array
X[:10], Y[:10]
```

```python
from sklearn.preprocessing import LabelEncoder #import preprocessing module  labelencoder use to Encodes categorical data into numerical values.

from tensorflow.keras.utils import to_categorical # Import from the correct location
```

```python
l_encode = LabelEncoder() # creates an instance of the LabelEncoder class from sklearn.preprocessing.
l_encode.fit(Y) # fits the LabelEncoder to the data in Y
Y = l_encode.transform(Y) #This line transforms the data in Y using the fitted LabelEncoder


Y = to_categorical(Y)
Y
```

```python
Y.shape #rows, col
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
scaled_X
#fit: Calculates the mean and standard deviation of each feature in the dataset X.
#transform: Applies the calculated mean and standard deviation to standardize the features in X.
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y = train_test_split(scaled_X,Y, test_size = 0.3, random_state = 1)
train_x.shape, train_y.shape, test_x.shape, test_y.shape
```

```python
train_x = train_x.reshape((train_x.shape[0], train_x.shape[1], 1))
test_x = test_x.reshape((test_x.shape[0], test_x.shape[1], 1)) #convert in 2D data
```

```python
from keras.models import Sequential #to import deep learning models
from keras.layers import Dense, Dropout
from keras.layers import Conv1D, Flatten # Import Flatten layer
import tensorflow as tf
```

```python

model = Sequential()
model.add(Conv1D(100, kernel_size=3, input_shape=(train_x.shape[1], 1))) #layer will produce 100 feature maps, layer will look at 3 consecutive values
model.add(Dropout(0.2)) #introduces randomness in the training process, helping to create a more generalized and robust model.
model.add(Flatten())  # Add a Flatten layer to convert the multi-dimensional output to 1D
model.add(Dense(train_y.shape[1], activation='softmax'))
```

```python
print(model.summary()) #dense- batch size, number of neuron, number_of_parameters = (input_size * output_size) + output_size
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])  #calculate error and accuracy
```

```python


# Define EarlyStopping callback
callback = tf.keras.callbacks.EarlyStopping(monitor = 'loss', patience = 3)

# Fit the model
history = model.fit(train_x, train_y, epochs = 50, batch_size = 32,
                    callbacks = [callback], validation_data=(test_x, test_y))
```

```python
model.save("Conv1D.keras")  # Save the model with the .keras extension
```

```python
plt.rcParams['figure.dpi'] = 800 #run tym config,number of pixel per inch
plt.plot(history.history['loss'])
plt.title('Conv1D model Train loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper right')
plt.savefig("LSTM_model_train_loss.pdf", format="pdf", bbox_inches="tight")
plt.show()

# include Test loss in graph
# check for exclusive another for test loss
```

```python
plt.rcParams['figure.dpi'] = 800
plt.plot(history.history['val_loss'])
plt.title('Conv1D model Validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper right')
plt.savefig("Conv1D_model_val_loss.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
scores = model.evaluate(test_x, test_y) #return loss and accuracy

for i, m in enumerate(model.metrics_names): #Iterates over the tuples generated by enumerate.
    print("\n%s: %.3f"% (m, scores[i])) # Prints the metric name and its corresponding value with three decimal places.
```

```python
predict_x=model.predict(test_x)
pred_=np.argmax(predict_x,axis=1) #find index of max value in each row

true_y = l_encode.inverse_transform(np.argmax(to_categorical(test_y[:]), axis = 1)[:,1]) #: This line converts the true labels test_y into their original categorical form

cnt = 0 # Initializes a counter cnt to 0.

for i,j in zip(pred_, true_y):
    print("Predicted: {}, True: {}".format(i, j)) # Prints the predicted and true labels for each data point.
    cnt = cnt+1 #Increments the counter cnt.


    if cnt > 100:
        break
```

```python
plt.rcParams['figure.dpi'] = 500
plt.plot(true_y, color = 'red', label = 'Real Class')
plt.plot(pred_, color = 'blue', label = 'Predicted Class')
plt.title('Classes of dataset')
plt.xlabel('Index')
plt.ylabel('Class')
plt.legend() # Creates a legend to differentiate the two plotted lines ("Real Class" and "Predicted Class").
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
rmse = np.sqrt(mean_squared_error(true_y, pred_))
mse = mean_squared_error(true_y, pred_)
mae = mean_absolute_error(true_y, pred_)
r2 = r2_score(true_y, pred_)
print("TEST RMSE:", rmse) #Root Mean Squared Error
print("TEST MSE:", mse) #Mean Squared Error)
print("TEST MAE:", mae) #Mean Absolute Error
print("TEST R2:", r2)
```
